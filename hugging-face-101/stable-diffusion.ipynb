## Load hugging face library and download all other model files 

```python
import torch
from diffusers import StableDiffusionPipeline, EulerDiscreteScheduler

model_id = "stabilityai/stable-diffusion-2"

# Use the Euler scheduler here instead
scheduler = EulerDiscreteScheduler.from_pretrained(model_id, subfolder="scheduler")
pipe = StableDiffusionPipeline.from_pretrained(model_id, scheduler=scheduler, torch_dtype=torch.float16)
pipe = pipe.to("cuda")
```

```python
# optionally you could save the model weights and can be used to load the model directly without downloading files over again 
```

```python
prompt = "a studio potrait photo with artificial lighting of a family looking outside a window"
image = pipe(prompt).images[0]
image
```

```python
image.save("family photo.png") # save the image to disk
```
